  For a reasoning model like DeepSeek-R1-7b, adding Chain-of-Thought (CoT) to the content field is critical.

  Here is why:

   1. Thinking Space: DeepSeek-R1 is architected to "think" before it acts. If you force it to output the JSON tool call immediately (with an empty content), you
      cripple its reasoning ability.
   2. Accuracy: By generating the CoT first (e.g., "I need to check the logs first, then search..."), the model conditions its own hidden state to generate the
      correct tool arguments.
   3. Debuggability: When the agent runs, you will see its internal monologue before it takes action, which is invaluable for debugging complex flows.
